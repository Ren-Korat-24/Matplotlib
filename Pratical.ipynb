19. Write a Python program to create bar plot of scores by group and 
gender. Use multiple X values on the same chart for men and women. 
Sample Data: 
Means (men) = (22, 30, 35, 35, 26) 
Means (women) = (25, 32, 30, 35, 29) 

```python
men=[20,30,35,35,26]
women=[25,32,30,35,29]
x=np.arange(5)
# x=np.average(men+women)
width=0.25

bars1 = plt.bar(x - width/2, men, width, label="Men")
bars2 = plt.bar(x + width/2, women, width, label="Women")

for i in range(len(men)):
    plt.text(x[i]-width/2,men[i],str(men[i]),label="Men",ha="center")
    plt.text(x[i]+width/2,women[i],str(women[i]),label="Women",ha="center")

plt.title("Scores by group and Gender")
plt.xlabel("Group")
plt.ylabel("Scores")

plt.show()
```

15. Write a Python program to create bar plots with error bars on the same 
figure 
Sample Date 
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824 
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645 

```python
Mean_velocity=[0.2474, 0.1235, 0.1737, 0.1824]
std_deviation = [0.3314, 0.2278, 0.2836, 0.2645]

x=np.arange(len(Mean_velocity))

plt.bar(x,Mean_velocity,yerr=std_deviation,capsize=5)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plotting Error Bars with plt.errorbar()')
plt.legend()
plt.grid(True)
plt.show()
```
